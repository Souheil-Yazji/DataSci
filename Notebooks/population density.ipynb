```python
import pandas as pd
from google.colab import files
```

#Load Dataset

```python
pop_dens = pd.read_csv('/content/pop_dens.csv')
pop_dens.head()
```

#Slice data 2014 to 2017

```python
temp_pop_dens = pop_dens.drop(pop_dens.columns.to_series()["Country Code":"2013"], axis=1)
sliced_pop_dens = temp_pop_dens.drop(temp_pop_dens.columns.to_series()["2018":], axis=1)

sliced_pop_dens.head()
```

#Remove NaNs

```python
cleaned_pop_dens = sliced_pop_dens.dropna()
cleaned_pop_dens["Country Name"]
```

#Save File

```python
cleaned_pop_dens.to_csv('pop_dens_2014_2017.csv')
```
